```python
# Import splinter and beautiful soup 
from splinter import Browser
from bs4 import BeautifulSoup as soup
import pandas as pd
```

```python
# Select WebBrowser for url scrapping
browser = Browser('edge')
```

```python
# Visit the statistics canada cancer site
url = 'https://www150.statcan.gc.ca/t1/tbl1/en/tv.action?pid=1310011101&pickMembers%5B0%5D=2.1&pickMembers%5B1%5D=3.1&pickMembers%5B2%5D=4.1&cubeTimeFrame.startYear=1992&cubeTimeFrame.endYear=2020&referencePeriods=19920101%2C20200101'
browser.visit(url)
```

```python
# Create a Beautiful Soup object
html = browser.html
html_soup = soup(html, 'html.parser')
```

```python
# Extract all the HTML elements within the body tag
body_elements = html_soup.body.find_all()
```

```python
# Extract all rows of data
table_rows = html_soup.find_all('tr')
```

```python
# Find the table element
table = html_soup.find('table')
```

```python
# Extract the table header (table column names)
header = table.find('thead')
```

```python
# Extract all the table header cells
header_cells = header.find_all('th')
```

```python
# Print the table header cells
for cell in header_cells:
    print(cell.text)
```

```python
# Extract the table body
body = table.find('tbody')
```

```python
# Extract all the rows within the table body
rows = body.find_all('tr')
```

```python
# Iterate over the rows and print their content
for row in rows:
    # Extract the cells within each row
    cells = row.find_all('td')
    for cell in cells:
        print(cell.text)
    print("---")
```

```python
# Create an empty list to store the table data
data = []
```

```python
# Iterate over the columns and extract their content
header = ['1992', '1993', '1994', '1995', '1996', '1997', '1998', '1999', '2000', '2001', '2002', '2003', '2004', '2005', '2006', '2007', '2008', '2009', '2010', '2011', '2012', '2013', '2014', '2015', '2016', '2017', '2018', '2019', '2020']

# Iterate over the rows and extract their content
for row in rows:
    # Extract the cells within each row
    cells = row.find_all('td')
    # Extract the text from each cell and append to the data list
    data_row = [cell.text for cell in cells]
    data.append(data_row)

# Create a data frame from the extracted table data
df = pd.DataFrame(data, columns=header)
```

```python
# Export Data into a csv
df.to_csv("CancerInCanada.csv")
```
