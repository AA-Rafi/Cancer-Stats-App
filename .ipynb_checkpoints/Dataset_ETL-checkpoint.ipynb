```python
# Import dependencies
import pandas as pd
import numpy as np
from pathlib import Path
```

```python
base_rates_df = pd.read_csv('Datasets/Number_and_rates_of_new_cases_of_primary_cancer_by_cancer_type_age_group_and_sex.csv', low_memory=False)
base_rates_df.head()
```

```python
remove_ratescol = ["DGUID", "UOM_ID", "VECTOR", "SCALAR_ID", "COORDINATE", "STATUS", "SYMBOL", "TERMINATED", "DECIMALS"]
rates_df= base_rates_df.drop(columns=remove_ratescol)
rates_df.tail()
```

```python
#export of rates
rates_df.to_csv('Datasets/rates.csv')
```
